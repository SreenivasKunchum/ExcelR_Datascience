```python
import pandas as pd
import numpy as np
```

```python
# Read csv file. Note that encoding changed to 'latin-1' since default 'utf-8' encoding is giving errors.
df_book = pd.read_csv('book.csv', encoding='latin-1')
df_book.head()
```

**Exploratory Data Analysis (EDA)**

```python
# Remove unnamed first column
df_book = df_book.iloc[:,1:]
df_book.head()
```

```python
df_book.sort_values('User.ID')
```

```python
# Number of unique users in the dataset
len(df_book['User.ID'].unique())
```

```python
# Number of unique books in the dataset
len(df_book['Book.Title'].unique())
```

```python
df_book['Book.Title'].value_counts()
```

```python
df_book[df_book[['User.ID', 'Book.Title']].duplicated()==True]
```

```python
# Remove Duplicates
#df_book[((df_book['User.ID'] == 3757) & (df_book['Book.Title'] == "The Magician's Tale"))]
df_book = df_book[df_book[['User.ID', 'Book.Title']].duplicated()==False]
```

```python
# After removing the duplicates, validating data for single instance
df_book[(df_book['User.ID']==278418) & (df_book['Book.Title'] == "White Oleander : A Novel")]
```

```python
df_users_books = df_book.pivot(index='User.ID', columns='Book.Title', values='Book.Rating')
df_users_books
```

```python
df_users_books = df_users_books.fillna(value=0)
df_users_books
```

```python
df_users_books.index
```

**Recommender System**

```python
# Calculating cosine similarities between the users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
#Note: Pairwise distances provides distances between two users. To get similarities, we need to subtract from 1
users_sim = 1 - (pairwise_distances(df_users_books.values, metric='cosine'))
users_sim 
```

```python
# Create a dataframe for above array
df_users_sim = pd.DataFrame(users_sim)
df_users_sim
```

```python
# Set the indexes and column names to user ids
df_users_sim.index = df_users_books.index
df_users_sim.columns = df_users_books.index
df_users_sim
```

```python
df_users_sim.iloc[0:5,0:5]
```

```python
# Diagonals are between the same user. So, fill them with zero.
np.fill_diagonal(users_sim, 0)
df_users_sim.iloc[0:5,0:5]
```

```python
# Most similar users
round(df_users_sim,6).idxmax()
```

```python
df_temp = df_users_sim[df_users_sim.index==278851]
round(df_temp[[278851,278202]],6)
```

```python
df_temp = df_users_sim[df_users_sim.index==10]
round(df_temp[[10,8]],6)
```
